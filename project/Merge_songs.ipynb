```python
import csv

# Open the first CSV file and read the rows into a list
with open('/home/ubuntu/songs.csv', newline='') as f1:
    reader1 = csv.reader(f1)
    rows1 = list(reader1)

# Open the second CSV file and read the rows into another list
with open('/home/ubuntu/songs_meta.csv', newline='') as f2:
    reader2 = csv.reader(f2)
    rows2 = list(reader2)

# Combine the rows from the two lists based on row number
combined_rows = [row1 + row2 for row1, row2 in zip(rows1, rows2)]

# Save the combined rows as a new CSV file
with open('songs_merge.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(combined_rows)
```
